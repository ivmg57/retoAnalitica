**LINEAR REGRESION**

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("peso_estatura_genero.csv")
```

```python
model = LinearRegression()
x = df['Estatura'].to_numpy(dtype='float').reshape((-1, 1))
y = df['Peso'].to_numpy(dtype='float')
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
b0 = model.intercept_
print(b0)
```

```python
b1 = model.coef_
print(b1)
```

```python
#y_ = b0 + b1*x
y_ = model.predict(x)
print(y_)
```

```python
fig = plt.figure()

plt.scatter(x,y,c='b',marker='*')
plt.plot(x,y_,c='r')
```
